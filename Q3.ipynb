How does awareness and understanding of the Online Safety Act vary across different age groups and digital literacy?

How familiar are you with the online safety act?

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data.csv" 
df = pd.read_csv(file_path)

familiarity_col = 'How familiar are you with the online safety act?'

familiarity_counts = df[familiarity_col].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(familiarity_counts.index, familiarity_counts.values, color='#ffb347')
plt.title("Familiarity with the Online Safety Act", fontsize=12)
plt.xlabel("Familiarity Level")
plt.ylabel("Number of Respondents")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()

print(familiarity_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data.csv"
df = pd.read_csv(file_path)

age_col = 'What is your age group?'
confidence_col = 'How would you rate your overall confidence using digital platforms? (e.g social media, e-banking, online forms)'

age_filtered_df = df[df[age_col].str.contains('18-24|18-25', na=False)]

confidence_counts = age_filtered_df[confidence_col].value_counts().sort_index()

plt.figure(figsize=(8, 5))
plt.bar(confidence_counts.index.astype(str), confidence_counts.values, color='#66b3ff')
plt.title("Confidence Using Digital Platforms Among Respondents Aged 1825", fontsize=12)
plt.xlabel("Confidence Rating (1 = Low, 5 = High)")
plt.ylabel("Number of Respondents")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print(confidence_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data.csv" 
df = pd.read_csv(file_path)

age_col = 'What is your age group?'
familiarity_col = 'How familiar are you with the online safety act?'

age_familiarity = pd.crosstab(df[age_col], df[familiarity_col])

age_familiarity.plot(kind='bar', figsize=(10, 6), colormap='tab20')
plt.title("Familiarity with the Online Safety Act by Age Group", fontsize=13)
plt.xlabel("Age Group")
plt.ylabel("Number of Respondents")
plt.legend(title="Familiarity Level", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print(age_familiarity)
```
