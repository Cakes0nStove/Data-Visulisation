<h1>How does awareness and understanding of the Online Safety Act vary across different age groups and digital literacy?</h1>

Saving Images Function

```python
# To plot pretty figures
import os
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "data_viz_group"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution, bbox_inches="tight")
```

<h3>How familiar are you with the online safety act?</h3>

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data.csv" 
df = pd.read_csv(file_path)

familiarity_col = 'How familiar are you with the online safety act?'

familiarity_counts = df[familiarity_col].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(familiarity_counts.index, familiarity_counts.values, color='#ffb347')
plt.title("Familiarity with the Online Safety Act", fontsize=12)
plt.xlabel("Familiarity Level")
plt.ylabel("Number of Respondents")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=30, ha='right')

save_fig("Q3.1_bar")
plt.tight_layout()
plt.show()

print(familiarity_counts)
```

<h3>How would you rate your overall confidence using digital platforms? (e.g social media, e-banking, online forms)</h3>

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data.csv"  
df = pd.read_csv(file_path)

confidence_col = 'How would you rate your overall confidence using digital platforms? (e.g social media, e-banking, online forms)'

confidence_counts = df[confidence_col].value_counts().sort_index()

plt.figure(figsize=(8, 5))
plt.bar(confidence_counts.index.astype(str), confidence_counts.values, color='#66b3ff')
plt.title("Overall Confidence Using Digital Platforms", fontsize=12)
plt.xlabel("Confidence Rating (1 = Low, 5 = High)")
plt.ylabel("Number of Respondents")
plt.grid(axis='y', linestyle='--', alpha=0.7)

save_fig("Q3.2_bar")
plt.tight_layout()
plt.show()

print(confidence_counts)
```

<h3>What is your age group?</h3>

```python

age_col = 'What is your age group?'
age_counts = df[age_col].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(
    age_counts.values,
    labels=age_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#ff9999', '#ffb347', '#ffd580', '#66b3ff', '#99ff99']
)
plt.title("Respondents by Age Group", fontsize=12)

save_fig("Q3.3_pie")
plt.tight_layout()
plt.show()
```

<h3> Relationship between digital literacy and familiarity with online safety act</h3>

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data.csv" 
df = pd.read_csv(file_path)

confidence_col = 'How would you rate your overall confidence using digital platforms? (e.g social media, e-banking, online forms)'
familiarity_col = 'How familiar are you with the online safety act?'

confident_users_df = df[df[confidence_col].isin([4, 5])]

familiarity_counts = confident_users_df[familiarity_col].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(
    familiarity_counts.values,
    labels=familiarity_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66b3ff', '#99ff99', '#ffcc99']
)
plt.title("Familiarity with the Online Safety Act (Among Confident Users  Ratings 4 & 5)", fontsize=12)

save_fig("Familiarity_Confident_Pie")
plt.tight_layout()
plt.show()

print(familiarity_counts)
```

<h3> Relationship between age and familiarity with online safety act</h3>

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data.csv" 
df = pd.read_csv(file_path)

age_col = 'What is your age group?'
familiarity_col = 'How familiar are you with the online safety act?'

age_familiarity = pd.crosstab(df[age_col], df[familiarity_col])

age_familiarity.plot(kind='bar', figsize=(10, 6), colormap='tab20')
plt.title("Familiarity with the Online Safety Act by Age Group", fontsize=13)
plt.xlabel("Age Group")
plt.ylabel("Number of Respondents")
plt.legend(title="Familiarity Level", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)

save_fig("Confidence_Age(All)_Bar")
plt.tight_layout()
plt.show()

print(age_familiarity)
```
